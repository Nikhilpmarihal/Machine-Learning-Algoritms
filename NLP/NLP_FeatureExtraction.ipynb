***BAG OF WORDS***

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'text':['People watch cricket','cricket watch cricket','people write comment','cricket write comment'],'output':[1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
```

```python
#VOCAB
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

```python
cv.transform(["Cricket watch and write comment of cricket"]).toarray()
```

***BAG OF N GRAM***

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(1,2))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
#VOCAB
print(len(cv.vocabulary_))
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

***TF-IDF***

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer()
tfidf = tf.fit_transform(df['text']).toarray()
```

```python
print(tf.idf_)
print(tf.get_feature_names_out())
```
