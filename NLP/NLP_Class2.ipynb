```python
import re
text = "The Apple is red. Pineapple is yellow . Apple!"
target_word = "Apple"

#/b ensures we match the whole word only
# re.IGNORECASE allows it to find "APPLE" , or "apple"
pattern = re.compile(r'\b' + re.escape(target_word) + r'\b', re.IGNORECASE)
# Substitute the match with its lowercase version
processed_text = pattern.sub(target_word.lower(),text)

print("Original text:", text)
print("Processed text:", processed_text)
```

```python
import re
text = "The Apple is red. The Carpet is blue. I like MANGO."
target_words = ["Apple", "car", "MANGO"]

#1.Create a pattern that joins words with OR
#\b ensures we match whole words only
pattern = re.compile(r'\b('+'|'.join([re.escape(w) for w in target_words]) + r')\b', re.IGNORECASE)
def lower_match(match):
  return match.group().lower()

processed_text = pattern.sub(lower_match, text)

print("Original text:", text)
print("Processed text:", processed_text)
```
